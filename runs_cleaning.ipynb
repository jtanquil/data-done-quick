## Extracting run categories

```python
runs['name_token'] = runs.Name.apply(lambda x: x.split(' '))
```

```python
runs.name_token.head()
```

```python
word_hist = {}

def word_count(tokens, word_hist):
    for token in tokens:
        if token in word_hist.keys():
            word_hist[token] += 1
        else:
            word_hist[token] = 1
            
    return word_hist
```

```python
runs.name_token.apply(lambda x: word_count(x, word_hist))
```

```python
runs[runs.Name.apply(lambda x: check_title(x, 'Donkey Kong Country'))]
```

```python
def tokenize(s):
    try:
        return s.split(' ')
    except:
        return []
```

```python
runs['desc_token'] = runs.Description.apply(tokenize)
```

```python
runs.desc_token.apply(lambda x: word_count(x, word_hist))
```

```python
words = pd.DataFrame.from_dict(word_hist, orient = 'index')
```

```python
words = words.rename(index = str, columns = {0 : 'count'})
```

```python
words.sort_values(by = 'count', ascending = False).head(10)
```

```python
words['word'] = words.index
```

check:
- all (all stages)
- no (no major glitches)
- new/ng (ng+)
- difficulty levels (easy/normal/hard + exceptions: insane, legendary, etc)

extraction: split after one of these words until end/./(others)

```python
# takes a string (description or name) and attempts to extract a coherent category
def extract_categories(s):
    # common delimiting words/strings for category names (going left to right)
    # TODO: convert these to regex for .findall
    categories = ['all', 'no', 'new', '\+', 'relay', 'race', 'TAS',
                  'oob', 'easy', 'normal', 'hard', 'insane', 'legendary', 
                  'ng\+']
    # capturing expressions for regex
    regex = ['\w+ ?% ', '\w+less', '\dp\dc', '\d players? \d controllers?', 
             '\d+[ -]exit|star|shine', 'in[- ]?bounds', 'new game ?\+', 'blind(fold)?', 
             '\d|multi|single[ -]?player', '(solo)? co-op']
    
    for exp in regex:
        cats = re.findall(exp, s)
        
        if len(cats) > 0:
            return cats[0]
```
