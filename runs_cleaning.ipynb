```python
word_hist = {}

def word_count(tokens, word_hist):
    for token in tokens:
        if token in word_hist.keys():
            word_hist[token] += 1
        else:
            word_hist[token] = 1
            
    return word_hist
```

```python
def search_string(s, x):
    try:
        return s.lower() in x.lower()
    except:
        return False

def search(s, df):
    return df.loc[(df.Description.apply(lambda x: search_string(s, x))) | (df.Name.apply(lambda x: search_string(s, x)))]
```

extraction: split after one of these words until end/./(others)

```python
# regex for common category names
regex = [r'\w+ ?%', r'\d+[ -](stars?|exits?|shines?)', '(^| )TAS(bot)?', r'\w+less', r'(^| )all [\w &]+',
         r'no [\w ]+', r'[\w -]+only', r'\w+ Mode', r'(true |dark )?\w* ?(?<!automata )ending( A)?', 'beat the game',
         'oob', '(in|out)[- ]?(of)?[- ]?bounds', r'(individual|bonus|secret|extra|master)?(?<!lost)( levels?| stages?)( \d)?( \w*)?',
         r'(community/ninja/retro)?\d*[- ]tracks?(?!er)( showcase)?', '(reverse boss order|RBO|reverse badge acquisition)', 'deathwarps',
         r'\dp\dc', r'\d (game|player)s? \d controllers?', 
         r'new game ?\+?', r'ng\+', 'blind(fold(ed)?)?', r'(\d|multi|single)[ -]?player', '(solo |^| )co-?op',
         '(relay)?(^| )race', '(1|one)[ -]?handed', 'bingo', r'\d+[- ](boss|meters|talisman|pups|cubes|loops|trinkets|egg|bags of ca\$h)',
         r'easy(\w| )*', r'normal(\w| )*', r'hard(est|core)?(\w| )*', r'insane(\w| )*', 
         r'beginner(\w| )*', r'advanced(\w| )*', r'expert(\w| )*',
         'legendary', 'tourist', 'not impossible', 'casual', 'brutal', 
         '(00 |perfect )?agent', '(european )?extreme', 'watch me die', 
         'heavy', 'ultra-violence', 'UV ', 'be gentle', r'\w* ?difficulty', 'pacifist', r'glitch(ed)?(\w| )+']
```

```python
def try_search(exp, x):
    try:
        return exp.search(x).group()
    except:
        return np.nan

def find_cat(s):
    exp = re.compile(s, re.IGNORECASE)
    return runs.Description.apply(lambda x: try_search(exp, x)), runs.Name.apply(lambda x: try_search(exp, x))
```

```python
runs.Description.str.findall(r'\d+[- ][a-zA-Z]+').str.join(' ').str.cat(sep = ' ')
```

```python
runs.Name.str.findall(r'\d+[- ][a-zA-Z]+').str.join(' ').str.cat(sep = ' ')
```

```python
def extract_categories(runs, regex):
    regex_cols = [exp + s for exp in regex for s in ['_desc', '_name']]
    
    for exp in regex:
        runs[exp + '_desc'], runs[exp + '_name'] = find_cat(exp)
    
    return runs

def concat_category(row, regex_cols):
    s = ''
    for col in regex_cols:
        try:
            if row[col].strip() not in s.strip():
                s += row[col] + ' '
        except:
            pass
    
    return s
```

```python
new_runs = extract_categories(runs.copy(), regex)
```

```python
regex_cols = [exp + s for exp in regex for s in ['_desc', '_name']]
```

```python
new_runs['category'] = new_runs.apply(lambda x: concat_category(x, regex_cols), axis = 1)
```

```python
# strips categories from game names by truncating the name after any word in the category appears
def strip_categories(row):
    # if there is no category, don't change the name
    if len(row.category) == 0:
        return row.Name
    else:
        # list of indices for the category words actually in the name
        indices = [row.Name.find(s) for s in row.category.split(' ') if len(s) > 0 and s in row.Name]
        
        if len(indices) == 0 or min(indices) == -1:
            return row.Name
        else:
            return row.Name[:min(indices)]
```

```python
new_runs['new_Name'] = new_runs.apply(strip_categories, axis = 1)
```

```python
new_runs = new_runs.drop(regex_cols, axis = 1)
```
