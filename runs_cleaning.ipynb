```python
word_hist = {}

def word_count(tokens, word_hist):
    for token in tokens:
        if token in word_hist.keys():
            word_hist[token] += 1
        else:
            word_hist[token] = 1
            
    return word_hist
```

extraction: split after one of these words until end/./(others)

```python
# regex for common category names
regex = [r'\w+ ?%', r'\d+[ -](stars?|exits?|shines?)', '(^| )TAS(bot)? ', r'\w+less', r'(^| |(?:[(]))all [\w &]+',
         r'no[- ](?!constancy)[\w ]+', r'[\w -]+only', r'\w+ Mode', r'(true |dark )?\w* (?!automata)?ending( A)?', 'beat the game', 
         r'(\d+|wily|assassin|amnesia)? ?\w+(?<!the) run(?![a-z])',
         'oob', '(in|out)[- ]?(of)?[- ]?bounds', r'(individual|bonus|secret|extra|master|custom)?(?<!lost)( levels?| stages?|lv)( |\d*)?( |\w*)?(data org rta)?',
         r'(200cc |community/ninja/retro)?(\d*|nitro)[- ]tracks?( 200cc items| showcase)?(?!er)', '(reverse boss order|RBO|reverse badge acquisition)', '(death)?warps?',
         r'(road to gehenna dlc|main|human|sentinels|undead)? ?campaigns?', 'staff ghosts', '(inner|fixed) world',
         r"\w*('s)?( and | or | vs[.]? )?\w*('s)?(?<!valdis)(?<!yoshi's)(?<!cave)(?<!toy)(?<!ball)(?<!ball 2)(?<!^A)(?<!Heart) story",
         r'\dp\dc', r'\d (game|player)s? \d controllers?', 'max speed', 'zero bosses', 'grand prix',
         r'new game ?\+?', r'(vergil)?( |\Z)ng\+?( devil hunter| plague)?', 'blind(fold(ed)?)?', r'(\d|multi|single)[ -]?(player|p)', '(solo |^| )co-?op',
         '(relay)?(^| )race', '(1|one)[ -]?handed', 'bingo', r'\d+[- ](boss|meters|talisman|pups|cubes|loops|trinkets|egg|characters?|bags of ca\$h|pokemon|episode|scroll)',
         r'easy(\w| )*', r'normal(\w| )*', r'hard(est|core)?(\w| )*', r'insane(\w| )*', 
         r'beginner(\w| )*', r'advanced(\w| )*', r'expert(\w| )*', ' nta', 'pro/sa',
         'legendary', 'tourist', 'not impossible', 'casual', 'brutal', 
         '(00 |perfect )?agent', '(european )?extreme', 'watch me die', 
         'heavy', 'ultra-violence', 'UV[- ]\w*', 'be gentle', r'\w* ?difficulty', 'pacifist', r'(no/low-)?glitch(ed)?(\w| )+', '(live tutorial )?(exhibition|showcase)']
# ((metroid prime|mega man|bad ending|\S*) (or|vs.))+ (metroid prime|good ending|\S*)
```

```python
def try_search(exp, x):
    try:
        return x[exp.search(x).start():exp.search(x).end()]
    except:
        return np.nan

def find_cat(s):
    exp = re.compile(s + r'[^.;*:()]*', re.IGNORECASE)
    return runs.Description.apply(lambda x: try_search(exp, x)), runs.Name.apply(lambda x: try_search(exp, x))

def compare_cats(current_cat, new_cat):
    try:
        if current_cat in new_cat:
            return new_cat
        elif new_cat in current_cat:
            return current_cat
        else:
            return current_cat + ', ' + new_cat
    except:
        if isinstance(current_cat, str):
            return current_cat
        elif isinstance(new_cat, str):
            return new_cat
        else:
            return ''
```

```python
def find_cats(regex):
    cats = pd.DataFrame({'desc_cat' : ['' for i in range(len(runs) + 2)], 'name_cat' : ['' for i in range(len(runs) + 2)]})
    
    for exp in regex:
        cats['desc_exp'], cats['name_exp'] = find_cat(exp)
        cats.desc_cat = cats[['desc_cat', 'desc_exp']].apply(lambda x: compare_cats(x.desc_cat, x.desc_exp), axis = 1)
        cats.name_cat = cats[['name_cat', 'name_exp']].apply(lambda x: compare_cats(x.name_cat, x.name_exp), axis = 1)
        
    return cats[['desc_cat', 'name_cat']].apply(lambda x: compare_cats(x.desc_cat, x.name_cat), axis = 1)
```

```python
# strips categories from game names by truncating the name after any word in the category appears
def strip_categories(row):
    # if there is no category, don't change the name
    if len(row.category) == 0:
        return row.Name
    else:
        # list of indices for the category words actually in the name
        indices = [row.Name.find(s) for s in row.category.split(' ') if len(s) > 0 and s in row.Name]
        
        if len(indices) == 0 or min(indices) == -1:
            return row.Name
        else:
            return row.Name[:min(indices)]
```

```python
new_runs['new_Name'] = new_runs.apply(strip_categories, axis = 1)
```
